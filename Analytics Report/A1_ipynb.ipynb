## First step: Reading files

As analysis starts with Economic stability analysis- we firstly search and read Datasets with economic indicators(World Bank dataset)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Firstly, we will be working with WORLD BANK FOLDER
#Readind all necessary files:

countries_continents=pd.read_csv("./World Bank/Countries_and_continents_of_the_world.csv")
age_struct = pd.read_csv("./World Bank/countries-dataset-2020/Coutries age structure.csv")
cost_of_liv_indx = pd.read_csv("./World Bank/countries-dataset-2020/Cost of living index by country 2020.csv")
quality_of_life_index = pd.read_csv("./World Bank/countries-dataset-2020/Quality of life index by countries 2020.csv")
household_exp = pd.read_csv("./World Bank/householdexpenditure/householdexpenditure.csv")
population_by_cntry= pd.read_csv('./World Bank/population-by-country-2020/population_by_country_2020.csv')
wdi_country= pd.read_csv('./World Bank/world-development-indicators/Country.csv')
```

## Second step: Preparing DataFrame to derive necessary information,
### Merging files for creating World Bank DataFrame

```python
#Now we start MERGING files to create DataFrame that can provide me with insights for Report
#This DataFrame will contain all necessary columns

#First, I will merge files in second folder
merged_countries_dataset = pd.merge(age_struct, cost_of_liv_indx, on="Country", how="left")
merged_countries_dataset = pd.merge(merged_countries_dataset, quality_of_life_index, on="Country", how="left")
#I will drop some of the columns that i dont need
merged_countries_dataset = merged_countries_dataset.drop(columns=['Restaurant Price Index', 'Safety Index', 'Health Care Index','Traffic Commute Time Index', 'Pollution Index' , 'Climate Index', 'Cost of Living Index_y','Local Purchasing Power Index', 'Groceries Index'])

#'Countries_and_continents_of_the_world' contains biggest number for countries- i will use this file as the main file,
#and will merge other files with this one
df_WB= pd.merge(countries_continents[['Country Name', 'Population', 'Continent', 'Development Status', 'Country Code',  'GNI per capita', 'GDP ($ per capita)',  'Region']], merged_countries_dataset, left_on="Country Name", right_on="Country", how="left")
#(5th folder) 'householdexpenditure'. Column 'country' has spaces- We will remove them before merging.
household_exp['Country'] = household_exp['Country'].str.strip()
df_WB= pd.merge(df_WB, household_exp[["Country", 'HouseholdExpenditurePerCapita']], left_on="Country Name", right_on="Country", how="left")
#(8th folder)'population-by-country-2020'
df_WB= pd.merge(df_WB, population_by_cntry[['Country (or dependency)', 'World Share']], left_on='Country Name', right_on= 'Country (or dependency)', how='left')
#(12th folder) 'world-development-indicators'
df_WB= pd.merge(df_WB, wdi_country[['ShortName','IncomeGroup']], left_on="Country Name", right_on="ShortName", how="left")


#I don't use files below in my World Bank DataFrame(some of them have same info., some of them contain unnecessary info.):
#(3rd folder) 'countries-of-the-world'. - I don't use it: I have all neccesary information that contains this file.
#(4th folder) 'environmental-variables-for-world-countries'- I don't use it.
#(6th folder) 'internet-users-in-africa'- I don't use it.
#(7th folder) 'oecd-unemployment-rate-2020'- I don't use it.
#(9th folder) 'unemployment-in-india'- I don't use it.
#(10th folder)'weo2020electricityaccessdatabase'- I don't use it.
#(11th folder)'world-countries-and-continents-details'- I don't use it.

#Now we wahe our WORLD BANK DataFrame. It contains columns that we do not need- we will trop these columns
df_WB = df_WB.drop(columns=['Country_x',"Country_y",'Country (or dependency)','ShortName'])
```

## Third step: Cleaning the Data

```python
#Now it is time for cleaning the Data- converting to proper Datatypes and removngg Null values.

#Before converting str values to float- I will remove unnecessary objects (like 'USD', commas, etc.)
df_WB['GNI per capita'] = df_WB['GNI per capita'].str.replace(" USD", "")
df_WB['GNI per capita'] = df_WB['GNI per capita'].str.replace(",", "")
df_WB['GNI per capita'] = df_WB['GNI per capita'].str.replace("--", "")
df_WB['World Share'] = df_WB['World Share'].str.replace(" %", "")
df_WB['World Share'] = df_WB['World Share'].str.replace("%", "")
df_WB['Age 0 to 14 Years'] = df_WB['Age 0 to 14 Years'].str.replace('%', '')
df_WB['Age 15 to 64 Years'] = df_WB['Age 15 to 64 Years'].str.replace('%', '')
df_WB['Age above 65 Years'] = df_WB['Age above 65 Years'].str.replace('%', '')

#converting into float
df_WB['GNI per capita'] = pd.to_numeric(df_WB['GNI per capita'], errors='coerce')
df_WB['Age 0 to 14 Years'] = df_WB['Age 0 to 14 Years'].astype(float)
df_WB['Age 15 to 64 Years'] = df_WB['Age 15 to 64 Years'].astype(float)
df_WB['Age above 65 Years'] = df_WB['Age above 65 Years'].astype(float)
df_WB['World Share'] = df_WB['World Share'].astype(float)


#Checking how many Null values I have in my world bank DataFrame
df_WB.isnull().sum(axis = 0)
# Filling them with the median column value
df_WB['Population'] = (df_WB['Population'].fillna(df_WB['Population'].median()))
df_WB['GNI per capita'] = (df_WB['GNI per capita'].fillna(df_WB['GNI per capita'].mean()))
df_WB['GDP ($ per capita)'] = (df_WB['GDP ($ per capita)'].fillna(df_WB['GDP ($ per capita)'].median()))
df_WB['Age 0 to 14 Years'] = (df_WB['Age 0 to 14 Years'].fillna(df_WB['Age 0 to 14 Years'].median()))
df_WB['Age 15 to 64 Years'] = (df_WB['Age 15 to 64 Years'].fillna(df_WB['Age 15 to 64 Years'].median()))
df_WB['Age above 65 Years'] = (df_WB['Age above 65 Years'].fillna(df_WB['Age above 65 Years'].median()))
df_WB['Cost of Living Index_x'] = (df_WB['Cost of Living Index_x'].fillna(df_WB['Cost of Living Index_x'].median()))
df_WB['Rent Index'] = (df_WB['Rent Index'].fillna(df_WB['Rent Index'].median()))
df_WB['Cost of Living Plus Rent Index'] = (df_WB['Cost of Living Plus Rent Index'].fillna(df_WB['Cost of Living Plus Rent Index'].median()))
df_WB['Quality of Life Index'] = (df_WB['Quality of Life Index'].fillna(df_WB['Quality of Life Index'].median()))
df_WB['Purchasing Power Index'] = (df_WB['Purchasing Power Index'].fillna(df_WB['Purchasing Power Index'].median()))
df_WB['Property Price to Income Ratio'] = (df_WB['Property Price to Income Ratio'].fillna(df_WB['Property Price to Income Ratio'].median()))
df_WB['HouseholdExpenditurePerCapita'] = (df_WB['HouseholdExpenditurePerCapita'].fillna(df_WB['HouseholdExpenditurePerCapita'].median()))
df_WB['World Share'] = (df_WB['World Share'].fillna(df_WB['World Share'].median()))
```

Before making final decision, it is important to take into consideration various cultural aspects to ensure successful integration into the new market. Australia and Qatar are two countries with distinct cultural characteristics that our company have to take into account when expanding into these markets.

#### Australia
Informality: Australian culture tends to be less formal compared to other cultures.

Multiculturalism: Australia is a diverse country with a large immigrant population, and our business will ahve to besensitive to the needs of diverse groups.

Direct communication: direct communication and a more upfront approach compared to other cultures.

#### Qatar:
Islam: Qatar is a Muslim country, need to be respectful of Islamic customs and practices, including dress codes and religious observances.

Hierarchy: Qatar has a hierarchical culture- respect for seniority and authority.

Gender roles: Qatar is a patriarchal society, and gender roles are distinct, with men and women having different roles and responsibilities.

Overall, businesses need to be aware of the cultural nuances of Australia and Qatar when expanding into these markets and make an effort to adapt their business practices and communication styles to be successful in these unique cultural contexts.
