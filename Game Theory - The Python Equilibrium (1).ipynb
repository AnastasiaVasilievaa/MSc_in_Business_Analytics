```python
from numpy import random
import pandas as pd

#introduction
print(f"""{'*'*80}
Hello, and Welcome to the game. Now, you will be faced with Game Theory Sutiation.
First of all, let me explain what does it mean. 
The Game Theory Sutiation is a situation where an individual must decide between
acting in their own self-interest or working towards the mutual benefit of all parties involved.

Imagine scenario where two employees at a company are requesting a salary raise, 
the management may use Game Theory to make a decision. The employees are separated
and asked to provide information about the other employee that the management is not aware of. 
If one employee reveals information,only their salary will increase, while if neither reveal 
anything, the raise will be split between them. However, if both employees reveal information, 
their salaries will remain unchanged. 


Here is the table for greater clarity.""")
#creating two tables: one with outcomes, one with points
print(f"""{'-'*70}""")
import pandas as pd
data = {"Your action": ["Reveal", "Reveal", "Silent", "Silent"],
        "Colleague's action": ["Reveal", "Silent", "Reveal", "Silent"],
        "Mgmt. dec.(1st empl)": ["no salary rise", "salary rise", "no salary rise", "split raise"],
        "Mgmt. dec.(2st empl)": ["no salary rise", "no salary rise", "salary rise", "split raise"] }
table1= pd.DataFrame(data)
print(table1)
print(f"""{'-'*70}""")

print(f"""
""")
print(f"""Here are the points that you can get throughout the game:
""")
print(f"""{'-'*50}""")

data = {"Your Action": ["Reveal", "Reveal", "Stay Silent", "Stay Silent"],
       "Opponent's Action": ["Reveal", "Stay Silent", "Reveal", "Stay Silent"],
       "Points": [3, 5, 0, 1]}

table = pd.DataFrame(data)
print(table)
print(f"""{'-'*50}""")
print(f"""Now you will be asked if you want to play against me(answer "yes"),
or you prefer to watch the game between 2 Python players(answer "no").  
""")

#defining variables
while True:
    answer1 = input(f"Please answer \"yes\" or \"no\" if you want to play with me this round. ").lower()
    if answer1.lower() == "yes":
        print("Okay! let\'s play, I will run a game between me and you. Let\'s see who will win.")        
    # defining variables for players' scores
        human_player_score = 0
        me_player_score = 0
    # defining the rewards
        rewards = {"rr": 3, "rs": 5, "sr": 0, "ss": 1}
    # run the game for 5 rounds
        for round in range(5):
            human_player_action = input("Now, please enter your action (r for reveal or s for stay silent): ").lower()
            while human_player_action != "r" and human_player_action != "s":
                print("Incorrect input, please enter 'r' or 's'.")
                human_player_action = input("Now, please enter your action (r for reveal or s for stay silent): ").lower()
            me_player_action = random.choice(["r", "s"]) 
            print(f"Hmmmm, I will choose : {me_player_action}")
        # now we update scores based on the rewards
            human_player_score += rewards[human_player_action + me_player_action]
            me_player_score += rewards[me_player_action + human_player_action]
            print(f"Your score is {human_player_score}")
            print(f"My score is {me_player_score}")
    # checking who the winner is after 5 rounds
        if human_player_score > me_player_score:
            print("Congratulations colleague, you got salary raise!")
        elif me_player_score > human_player_score:
            print("I got salary raise, but don\'t be upset, keep working, maybe next nime you will be more lucky;)")  
        else:
            print("It\'s a draw!")
        break
    elif answer1.lower() == "no":
        print("Okay, then I will run a game between 2 Python players, there will be 5 rounds. You will just watch.")
    # defining variables for players' scores (game between 2 python players)
        player1_score = 0
        player2_score = 0
    # defining rewards
        rewards = {"rr": 3, "rs": 5, "sr": 0, "ss": 1}
    # run the game for a certain number of rounds
        for round in range(5):
            player1_action = random.choice(["r", "s"])
            if random.random() < 0.3:
                player2_action = "s"
            else:
                player2_action = "r"
        # updating scores based on rewards 
            player1_score += rewards[player1_action + player2_action]
            player2_score += rewards[player2_action + player1_action]
            print(f"1st employee score is {player1_score}")
            print(f"2nd employee's 2 score is {player2_score}")
        if player1_score > player2_score:
            print("After 5 rounds, employee 1 won and got salary raise!")
        elif player2_score > player1_score:
            print("After 5 rounds, employee 2 won and got salary raise!")  
        else:
            print("It is a draw!")
        break
    else:
        print("Your input is invalid. Please enter yes or no.")
        
print("""
""")
print(f"""{'*'*80}""")
```
